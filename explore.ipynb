```python
import pandas as pd
import os
```

```python
data_path=os.listdir("dataset/dynamic")

file_to_key = {
    "olist_sellers_dataset.csv": "olist_sellers_dataset",
    "product_category_name_translation.csv": "product_category_name_translation",
    "olist_orders_dataset.csv": "olist_orders_dataset",
    "olist_order_items_dataset.csv": "olist_order_items_dataset",
    "olist_customers_dataset.csv": "olist_customers_dataset",
    "olist_geolocation_dataset.csv": "olist_geolocation_dataset",
    "olist_order_payments_dataset.csv": "olist_order_payments_dataset",
    "olist_order_reviews_dataset.csv": "olist_order_reviews_dataset",
    "olist_products_dataset.csv": "olist_products_dataset"
}

datasets={}

for path in data_path:
    if path in file_to_key:
            key=path[:-4]
            if key not in datasets:
                  datasets[key]=pd.read_csv(os.path.join("dataset/dynamic",path))

print(datasets.keys())
```

```python
all_df_name=list(datasets.keys())
for df_name in all_df_name:
    print(""+60*"="+df_name+"="*60)
    print(datasets[df_name].info())
    print(datasets[df_name].dtypes)
    # print(datasets[df_name].head())
```

```python
df=datasets[all_df_name[2]]
drop_df = df[(df["order_approved_at"].isna()) & (df["order_status"] == "delivered")].index
datasets[all_df_name[2]].drop(drop_df,inplace=True)

df=datasets[all_df_name[7]]
df.loc[(df["review_comment_message"].isna()) & ~(df["review_comment_title"].isna()),"review_comment_message"]="No Comment!"
df.loc[~(df["review_comment_message"].isna()) & (df["review_comment_title"].isna()),"review_comment_title"]="No Title"

```
